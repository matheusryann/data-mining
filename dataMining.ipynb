```python
poluentes_unicos = df['Pollutant'].unique()
print(poluentes_unicos)
```

```python
# Dicion√°rio de massas molares dos poluentes (g/mol)
massa_molar = {
    'PM2.5': None,     
    'CO': 28.01,       
    'SO2': 64.06,      
    'PM10': None,      
    'NO2': 46.0055,    
    'O3': 48.00,       
    'NO': 30.006,      
    'NOX': None,       
    'BC': None,        
    'PM1': None        
}

#Fun√ß√£o para padroniza√ß√£o dos dados da coluna "value", transformando os dados de ppm para ¬µg/m¬≥
def convert_to_ug_m3(value, unit, pollutant):
    R = 0.0821  # Constante dos gases ideais (L¬∑atm/(K¬∑mol))
    T = 298  # Temperatura padr√£o (K)
    P = 1  # Press√£o padr√£o (atm)
    
    if unit == 'ppm':
        if pollutant in massa_molar and massa_molar[pollutant] is not None:
            return value * massa_molar[pollutant] * (P / (R * T)) * 1000  # Multiplica por 1000 para ¬µg/m¬≥
        else:
            return value
    elif unit == '¬µg/m¬≥':
        return value  
    else:
        return None  
        
df['Value_converted'] = df.apply(
    lambda row: convert_to_ug_m3(row['Value'], row['Unit'], row['Pollutant']),
    axis=1
)


print(df[['Pollutant', 'Unit', 'Value', 'Value_ug_m3']].head(20))
```

```python
print(df['Value_converted'].describe())
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Value_converted'])
plt.title("Boxplot dos Valores Convertidos")
plt.show()
```

```python
# Calcular os limites baseados no IQR
q1 = df['Value_converted'].quantile(0.25)
q3 = df['Value_converted'].quantile(0.75)
iqr = q3 - q1
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

# Filtrar os dados para remover outliers
df1 = df[(df['Value_converted'] >= limite_inferior) & (df['Value_converted'] <= limite_superior)]

# Exibir o n√∫mero de registros ap√≥s remo√ß√£o
print(f"N√∫mero de registros ap√≥s remo√ß√£o de outliers: {len(df1)}")
```
