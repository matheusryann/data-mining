```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Nome do arquivo (j√° no diret√≥rio do Jupyter)
file_name = "openaq.csv"

# Carregar o arquivo com o delimitador correto (ponto e v√≠rgula)
try:
    df = pd.read_csv(file_name, delimiter=";", encoding="utf-8")
    print("‚úÖ Arquivo carregado com sucesso!")
except Exception as e:
    print(f"Erro ao carregar o arquivo: {e}")

# Exibir as primeiras 5 linhas para verificar os dados
display(df.head())


```

```python
print("\nüîç √öltimas 5 linhas do dataset:")
display(df.tail())
```

```python

# Vis√£o geral do dataset

print("\nüìå Informa√ß√µes Gerais do Dataset:")
df.info()
```

```python
# Mostrar tipo de dados de cada coluna
print("\nüìå Tipos de dados de cada coluna:")
print(df.dtypes)
```

```python
# Retorna as dimens√µes do DataFrame (linhas, colunas)
print("\nüìå Dimens√µes do DataFrame:", df.shape)
```

```python
# Estat√≠sticas descritivas
print("\nüìä Estat√≠sticas descritivas:")
display(df.describe())
```

```python
# Verificar valores nulos
print("\nüîç Quantidade de valores nulos por coluna:")
print(df.isnull().sum())
```

```python
# Contagem de valores n√£o nulos por coluna
print("\nüìå Contagem de valores N√ÉO nulos por coluna:")
print(df.notnull().sum())
```

```python
# Identificar colunas num√©ricas e categ√≥ricas
print("\nüìå Colunas num√©ricas:")
print(df.select_dtypes(include=["number"]).columns)
```

```python
# Contagem de valores √∫nicos por coluna
print("\nüîç Quantidade de valores √∫nicos por coluna:")
print(df.nunique())
```

```python
# Verificar duplicatas ar no DataFrame
duplicatas = df.duplicated().sum()
print("\nüîç N√∫mero de registros duplicados:", duplicatas)
```

```python
# Separar 'Coordinates' em 'Latitude' e 'Longitude'
if 'Coordinates' in df.columns:
    df[['Latitude', 'Longitude']] = df['Coordinates'].str.split(',', expand=True)
    df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
    df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')
    print("‚úÖ Colunas 'Latitude' e 'Longitude' criadas com sucesso!")
    display(df[['Coordinates', 'Latitude', 'Longitude']].head())
```

```python
# Converter 'Last Updated' para datetime
if 'Last Updated' in df.columns:
    df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
    print("‚úÖ Coluna 'Last Updated' convertida para datetime com sucesso!")
    display(df[['Last Updated']].head())
```

```python
# Histograma para distribui√ß√£o de 'Value'

plt.figure(figsize=(10, 6))
sns.histplot(df['Value'], bins=50, kde=True, color='blue')
plt.title("Distribui√ß√£o da Coluna 'Value'", fontsize=14)
plt.xlabel("Value")
plt.ylabel("Frequ√™ncia")
plt.show()
```

```python
# Boxplot para identificar outliers em 'Value'

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Value'], color='orange')
plt.title("Boxplot da Coluna 'Value'", fontsize=14)
plt.xlabel("Value")
plt.show()
```

```python
# Gr√°fico de dispers√£o geogr√°fica

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', data=df, alpha=0.6)
plt.title("Dispers√£o Geogr√°fica (Latitude vs Longitude)", fontsize=14)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
